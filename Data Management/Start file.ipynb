```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url  = "https://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
title = soup.title.text
print(title)
```

```python
links = soup.find_all('a')
```

```python
#Print the first 10 rows
rows = soup.find_all('tr')
```

```python
for row in rows:
    row_td = row.find_all('td')
    print(row_td)
```

```python
str_cells = str(row_td)
print(str_cells)
```

```python
cleantext = BeautifulSoup(str_cells, 'lxml').get_text()
print(cleantext)
```

```python
import re
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'',str_cells))
    list_rows.append(clean2)
```

```python
df = pd.DataFrame(list_rows)
```

```python
df.head(10)
```

```python
df.head()
```

```python
df3 = df[0].str.split(',', expand = True)
df3.head(10)
```

```python
all_headers = []
col_str = str(col_label)
cleanlabel = BeautifulSoup(col_str, 'lxml').get_text()
all_headers.append(cleanlabel)
print(all_headers)
```
